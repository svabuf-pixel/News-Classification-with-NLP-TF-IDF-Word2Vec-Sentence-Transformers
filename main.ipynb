```python
import re
import csv
import nltk
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
from collections import Counter
from nltk.corpus import stopwords, wordnet
from nltk.stem import WordNetLemmatizer, PorterStemmer
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.tag import pos_tag
from sklearn.model_selection import train_test_split, StratifiedKFold, cross_val_score
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import LinearSVC
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import Pipeline
from sklearn.metrics import (
    accuracy_score, f1_score, confusion_matrix, ConfusionMatrixDisplay,
    classification_report, cohen_kappa_score,
    roc_auc_score, RocCurveDisplay, PrecisionRecallDisplay
)
from sklearn.multiclass import OneVsRestClassifier
from sklearn.preprocessing import label_binarize
from sklearn.decomposition import PCA
from mpl_toolkits.mplot3d import Axes3D
from gensim.models import Word2Vec
from sentence_transformers import SentenceTransformer

# =====================================================================
# 1. SETUP & DOWNLOADS
# =====================================================================
print("Initializing environment...")

REQUIRED_NLTK = {
    'tokenizers/punkt_tab':                  'punkt_tab',
    'corpora/stopwords':                     'stopwords',
    'corpora/wordnet':                       'wordnet',
    'taggers/averaged_perceptron_tagger_eng':'averaged_perceptron_tagger_eng',
}
for find_path, download_id in REQUIRED_NLTK.items():
    try:
        nltk.data.find(find_path)
    except LookupError:
        print(f"  Downloading NLTK resource: {download_id}")
        nltk.download(download_id, quiet=True)

try:
    STOP_WORDS = set(stopwords.words('english'))
except LookupError:
    STOP_WORDS = {
        'a','an','and','are','as','at','be','been','but','by','for',
        'from','has','have','he','in','is','it','its','of','on','or',
        'she','that','the','their','they','this','to','was','we','will',
        'with','you','your','i','me','my','our','ours','them','his','her'
    }


# =====================================================================
# 2. HELPER: FULL EVALUATION REPORT
# =====================================================================
# Centralised function so every model gets the same treatment.
# Prints: classification report, Cohen's Kappa, ROC-AUC (OvR macro),
# Confusion Matrix, and a per-class PR curve plot.

def evaluate_model(
    name: str,
    y_true,
    y_pred,
    decision_scores=None,   # required for ROC/PR curves (shape n_samples × n_classes)
    classes=None,
    cm_cmap=plt.cm.Blues,
):
    """
    Full evaluation suite for a single model.
    Returns a dict with scalar metrics for the final comparison table.
    """
    print(f"\n{'='*60}")
    print(f"  EVALUATION: {name}")
    print(f"{'='*60}")

    acc   = accuracy_score(y_true, y_pred)
    f1    = f1_score(y_true, y_pred, average='weighted')
    kappa = cohen_kappa_score(y_true, y_pred)

    print(f"  Accuracy      : {acc*100:.2f}%")
    print(f"  Weighted F1   : {f1*100:.2f}%")
    print(f"  Cohen's Kappa : {kappa:.4f}  "
          f"({'almost perfect' if kappa>0.8 else 'substantial' if kappa>0.6 else 'moderate' if kappa>0.4 else 'fair'})")

    print(f"\n--- Per-class Classification Report ---")
    print(classification_report(y_true, y_pred))

    # --- Confusion Matrix ---
    cm = confusion_matrix(y_true, y_pred)
    fig, ax = plt.subplots(figsize=(8, 6))
    ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=classes).plot(
        cmap=cm_cmap, ax=ax
    )
    ax.set_title(f'Confusion Matrix — {name}', fontsize=13, fontweight='bold')
    plt.tight_layout()
    plt.show()

    # --- ROC & PR Curves (only when decision scores are available) ---
    auc_macro = None
    if decision_scores is not None and classes is not None:
        n_classes = len(classes)
        is_binary = n_classes == 2

        # label_binarize returns shape (n_samples,) for binary — must reshape to 2-D
        # so the rest of the code can index with [:, i] uniformly.
        y_bin = label_binarize(y_true, classes=classes)
        if is_binary:
            y_bin = np.hstack([1 - y_bin, y_bin])  # shape (n, 2)

        # decision_function returns shape (n_samples,) for binary LinearSVC —
        # expand to (n, 2) as [score_class0, score_class1] with sign flipped for class 0.
        dec = np.array(decision_scores)
        if dec.ndim == 1:
            dec = np.vstack([-dec, dec]).T  # shape (n, 2)

        fig, axes = plt.subplots(1, 2, figsize=(14, 5))
        fig.suptitle(f'ROC & Precision-Recall Curves — {name}', fontsize=13, fontweight='bold')

        # ROC curves
        for i, cls in enumerate(classes):
            RocCurveDisplay.from_predictions(
                y_bin[:, i], dec[:, i],
                name=f"Class {cls}", ax=axes[0]
            )
        axes[0].plot([0, 1], [0, 1], 'k--', lw=1)
        axes[0].set_title('ROC Curves (One-vs-Rest)')
        axes[0].legend(fontsize=8)

        # Precision-Recall curves
        for i, cls in enumerate(classes):
            PrecisionRecallDisplay.from_predictions(
                y_bin[:, i], dec[:, i],
                name=f"Class {cls}", ax=axes[1]
            )
        axes[1].set_title('Precision-Recall Curves (One-vs-Rest)')
        axes[1].legend(fontsize=8)

        plt.tight_layout()
        plt.show()

        # Macro ROC-AUC
        try:
            if is_binary:
                auc_macro = roc_auc_score(y_bin[:, 1], dec[:, 1])
            else:
                auc_macro = roc_auc_score(y_bin, dec, average='macro', multi_class='ovr')
            print(f"  ROC-AUC (macro OvR) : {auc_macro:.4f}")
        except (ValueError, TypeError) as e:
            print(f"  ROC-AUC could not be computed ({type(e).__name__}): {e}")

    return {'model': name, 'accuracy': acc, 'f1_weighted': f1, 'kappa': kappa, 'roc_auc': auc_macro}


# =====================================================================
# 3. HELPER: ERROR ANALYSIS
# =====================================================================

def error_analysis(name: str, texts, y_true, y_pred, n_samples: int = 10):
    """
    Print a sample of misclassified examples and text-length distribution
    for correct vs incorrect predictions.
    """
    print(f"\n--- Error Analysis: {name} ---")
    results = pd.DataFrame({
        'text':      texts.values if hasattr(texts, 'values') else texts,
        'true':      y_true.values if hasattr(y_true, 'values') else y_true,
        'predicted': y_pred,
    })
    errors = results[results['true'] != results['predicted']].copy()
    correct = results[results['true'] == results['predicted']].copy()

    print(f"  Total errors : {len(errors)} / {len(results)} "
          f"({len(errors)/len(results)*100:.1f}%)")

    # Text length comparison
    errors['text_len']   = errors['text'].str.split().str.len()
    correct['text_len']  = correct['text'].str.split().str.len()
    print(f"  Avg words — correct: {correct['text_len'].mean():.1f}  |  "
          f"misclassified: {errors['text_len'].mean():.1f}")

    # Most common confusion pairs
    confusion_pairs = (
        errors.groupby(['true','predicted'])
              .size()
              .reset_index(name='count')
              .sort_values('count', ascending=False)
    )
    print(f"\n  Top confusion pairs (true → predicted):")
    print(confusion_pairs.head(5).to_string(index=False))

    # Sample misclassified texts
    print(f"\n  Sample misclassified (up to {n_samples}):")
    sample = errors.head(n_samples)
    for _, row in sample.iterrows():
        snippet = str(row['text'])[:120].replace('\n', ' ')
        print(f"    [true={row['true']} | pred={row['predicted']}] {snippet}")


# =====================================================================
# 4. PREPROCESSOR CLASSES
# =====================================================================

class BasePreprocessor:
    CONTRACTIONS = {
        "don't":"do not",       "doesn't":"does not",   "didn't":"did not",
        "can't":"cannot",       "couldn't":"could not",  "won't":"will not",
        "wouldn't":"would not", "isn't":"is not",        "aren't":"are not",
        "wasn't":"was not",     "weren't":"were not",    "hasn't":"has not",
        "haven't":"have not",   "hadn't":"had not",
        "I've":"I have",        "i've":"I have",
        "you've":"you have",    "we've":"we have",       "they've":"they have",
        "I'm":"I am",           "i'm":"I am",
        "you're":"you are",     "he's":"he is",          "she's":"she is",
        "it's":"it is",         "we're":"we are",        "they're":"they are",
        "I'll":"I will",        "i'll":"I will",
        "you'll":"you will",    "it'll":"it will",
    }

    def __init__(self):
        self.lemmatizer = WordNetLemmatizer()
        keys = sorted(self.CONTRACTIONS, key=len, reverse=True)
        self._contraction_pattern = re.compile(
            r'\b(' + '|'.join(re.escape(k) for k in keys) + r')\b',
            flags=re.IGNORECASE,
        )

    def expand_contractions(self, text: str) -> str:
        def replace(match):
            m = match.group(0)
            return self.CONTRACTIONS.get(m, self.CONTRACTIONS.get(m.lower(), m))
        return self._contraction_pattern.sub(replace, text)

    def _get_wordnet_pos(self, tag: str) -> str:
        if tag.startswith('J'): return wordnet.ADJ
        if tag.startswith('V'): return wordnet.VERB
        if tag.startswith('N'): return wordnet.NOUN
        if tag.startswith('R'): return wordnet.ADV
        return wordnet.NOUN


class TextPreprocessorTFIDF(BasePreprocessor):
    def __init__(self, lowercase=True, remove_stopwords=True, use_lemmatize=True, min_token_len=2):
        super().__init__()
        self.lowercase        = lowercase
        self.remove_stopwords = remove_stopwords
        self.use_lemmatize    = use_lemmatize
        self.min_token_len    = min_token_len
        self.stemmer          = PorterStemmer()

    def remove_noise(self, text: str) -> str:
        text = re.sub(r'https?://\S+|www\.\S+', '', text)
        text = re.sub(r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,}\b', '', text)
        text = re.sub(r'[@#]\w+', '', text)
        text = re.sub(r'[^a-zA-Z\s]', ' ', text)
        text = re.sub(r'\s+', ' ', text).strip()
        return text

    def process(self, text: str) -> list:
        text = self.expand_contractions(text)
        text = self.remove_noise(text)
        if self.lowercase: text = text.lower()
        tokens = word_tokenize(text)
        tagged = pos_tag(tokens)
        if self.use_lemmatize:
            tokens = [self.lemmatizer.lemmatize(t, pos=self._get_wordnet_pos(tag)) for t, tag in tagged]
        else:
            tokens = [self.stemmer.stem(t) for t, _ in tagged]
        if self.remove_stopwords:
            tokens = [t for t in tokens if t not in STOP_WORDS]
        return [t for t in tokens if len(t) >= self.min_token_len]


class TextPreprocessorWord2Vec(BasePreprocessor):
    def __init__(self, lowercase=True, remove_stopwords=True, min_token_len=2):
        super().__init__()
        self.lowercase        = lowercase
        self.remove_stopwords = remove_stopwords
        self.min_token_len    = min_token_len

    def clean_text(self, text: str) -> str:
        text = re.sub(r'https?://\S+|www\.\S+', '', text)
        text = re.sub(r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,}\b', '', text)
        text = re.sub(r'[@#]\w+', '', text)
        text = re.sub(r'[^a-zA-Z\s\.\!\?]', ' ', text)
        text = re.sub(r'\s+', ' ', text).strip()
        return text

    def process_sentence(self, sentence_str: str) -> list:
        if self.lowercase: sentence_str = sentence_str.lower()
        tokens = word_tokenize(sentence_str)
        tagged = pos_tag(tokens)
        clean  = []
        for word, tag in tagged:
            if not re.match(r'[a-zA-Z]', word): continue
            lemma = self.lemmatizer.lemmatize(word, pos=self._get_wordnet_pos(tag))
            if self.remove_stopwords and lemma in STOP_WORDS: continue
            if len(lemma) < self.min_token_len: continue
            clean.append(lemma)
        return clean

    def process_corpus(self, text: str) -> list:
        text = self.expand_contractions(text)
        text = self.clean_text(text)
        return [
            s for sent in sent_tokenize(text)
            if len(s := self.process_sentence(sent)) >= 2
        ]


class TextPreprocessorTransformer():
    def __init__(self, lowercase=False):
        super().__init__()
        self.lowercase = lowercase

    def clean_text(self, text: str) -> str:
        text = re.sub(r'https?://\S+|www\.\S+', '', text)
        text = re.sub(r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,}\b', '', text)
        text = re.sub(r'\s+', ' ', text).strip()
        return text

    def process(self, text: str) -> str:
        #text = self.expand_contractions(text)
        text = self.clean_text(text)
        if self.lowercase: text = text.lower()
        return text


# =====================================================================
# 5. DATA LOADING & PREPARATION
# =====================================================================
print("Loading datasets...")
try:
    df_training = pd.read_csv("dataset/training_data.csv", header=None)
    df_test     = pd.read_csv("dataset/testing_data.csv",  header=None)
except FileNotFoundError:
    print("Files not found — using dummy data.")
    df_training = pd.DataFrame({0: [
        "1 I love learning NLP", "1 Deep learning is fascinating", "1 Neural networks are powerful",
        "1 Python is great for data science", "1 Transformers changed NLP forever",
        "1 Word embeddings capture semantics", "1 BERT is a pretrained language model",
        "1 Machine learning helps automate tasks", "1 Gradient descent optimises weights",
        "1 Attention mechanisms improve translation",
        "2 The economy is shifting", "2 Stock market crashes today",
        "2 Inflation affects purchasing power", "2 Central banks raise interest rates",
        "2 GDP growth slowed last quarter", "2 Oil prices surge amid conflict",
        "2 Unemployment falls to record low", "2 Federal Reserve signals rate hike",
        "2 Trade deficit widens unexpectedly", "2 Housing market shows signs of cooling",
    ]})
    df_test = pd.DataFrame({0: ["1 I like coding", "2 The stock market rallied"]})

df_training.rename(columns={df_training.columns[0]: "news_column"}, inplace=True)
df_test.rename(    columns={df_test.columns[0]:     "news_column"}, inplace=True)
df_training['label'] = df_training['news_column'].str[0].astype(int)

# ── Initialise preprocessors ──────────────────────────────────────────
preprocessortfidf = TextPreprocessorTFIDF()
preprocessorw2v   = TextPreprocessorWord2Vec()
preprocessortrans = TextPreprocessorTransformer()

print("Cleaning text for TF-IDF...")
df_training['tokens_tfidf']       = df_training['news_column'].apply(lambda x: preprocessortfidf.process(str(x)[1:]))
df_training['cleaned_text_tfidf'] = df_training['tokens_tfidf'].apply(lambda x: " ".join(x))
df_test['cleaned_text_tfidf']     = df_test['news_column'].apply(lambda x: " ".join(preprocessortfidf.process(str(x)[1:])))

print("Cleaning text for Word2Vec...")
df_training['tokens_w2v']      = df_training['news_column'].apply(lambda x: preprocessorw2v.process_corpus(str(x)[1:]))
df_training['tokens_w2v_flat'] = df_training['tokens_w2v'].apply(lambda x: [t for s in x for t in s])
df_test['tokens_w2v_flat']     = df_test['news_column'].apply(lambda x: [t for s in preprocessorw2v.process_corpus(str(x)[1:]) for t in s])

print("Cleaning text for Sentence Transformers...")
df_training['cleaned_text_trans'] = df_training['news_column'].apply(lambda x: preprocessortrans.process(str(x)[1:]))
df_test['cleaned_text_trans']     = df_test['news_column'].apply(lambda x: preprocessortrans.process(str(x)[1:]))


# =====================================================================
# 6. DATA VISUALIZATION
# =====================================================================
print("\n--- Class Distribution ---")
class_counts = df_training['label'].value_counts(normalize=True).sort_index()
print(class_counts.apply(lambda x: f"{x*100:.1f}%"))

IMBALANCE_THRESHOLD = 0.15  # flag if any class holds < 15% of samples
is_imbalanced = class_counts.min() < IMBALANCE_THRESHOLD
if is_imbalanced:
    print("⚠️  Imbalanced dataset detected — class_weight='balanced' will be used in all classifiers.")

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Class distribution bar
class_counts_abs = df_training['label'].value_counts().sort_index()
axes[0].bar(class_counts_abs.index.astype(str), class_counts_abs.values, color='steelblue')
axes[0].set_title('Class Distribution', fontweight='bold')
axes[0].set_xlabel('Class')
axes[0].set_ylabel('Count')
for i, v in enumerate(class_counts_abs.values):
    axes[0].text(i, v + 0.5, str(v), ha='center', fontsize=9)

# Top-20 tokens
all_words   = [w for toks in df_training['tokens_tfidf'] for w in toks]
top_20      = Counter(all_words).most_common(20)
words, cnts = zip(*top_20)
axes[1].barh(words, cnts, color='coral')
axes[1].set_title('Top 20 Processed Tokens (TF-IDF)', fontweight='bold')
axes[1].invert_yaxis()

plt.tight_layout()
plt.show()


# =====================================================================
# 7. SHARED TRAIN / VAL SPLIT  (same indices for all models)
# =====================================================================
CLASS_WEIGHT = 'balanced' if is_imbalanced else None
CLASSES      = sorted(df_training['label'].unique())

train_idx, val_idx = train_test_split(
    df_training.index, test_size=0.2, random_state=42, stratify=df_training['label']
)


# =====================================================================
# 8. MODEL 1: TF-IDF + LinearSVC  (with baselines & cross-validation)
# =====================================================================
print("\n" + "="*60)
print("  MODEL 1: TF-IDF + LinearSVC")
print("="*60)

X_train_tf = df_training.loc[train_idx, 'cleaned_text_tfidf']
X_val_tf   = df_training.loc[val_idx,   'cleaned_text_tfidf']
y_train_tf = df_training.loc[train_idx, 'label']
y_val_tf   = df_training.loc[val_idx,   'label']

# ── Baseline comparisons ──────────────────────────────────────────────
print("\n  [Baselines]")
for name, clf in [
    ("LogisticRegression", LogisticRegression(max_iter=1000, class_weight=CLASS_WEIGHT, random_state=42)),
    ("MultinomialNB",      None),   # handled below — NB needs non-negative TF-IDF
]:
    pipe = Pipeline([
        ('tfidf', TfidfVectorizer(max_features=25000, ngram_range=(1,2), sublinear_tf=True)),
        ('clf',   clf if clf else MultinomialNB()),
    ])
    pipe.fit(X_train_tf, y_train_tf)
    yp = pipe.predict(X_val_tf)
    print(f"    {name:25s} | Acc: {accuracy_score(y_val_tf, yp)*100:.2f}%  "
          f"F1: {f1_score(y_val_tf, yp, average='weighted')*100:.2f}%")

# ── Main TF-IDF pipeline ──────────────────────────────────────────────
tfidf_pipeline = Pipeline([
    ('tfidf',      TfidfVectorizer(max_features=25000, ngram_range=(1,2), sublinear_tf=True)),
    ('classifier', LinearSVC(C=1.0, class_weight=CLASS_WEIGHT, random_state=42, max_iter=2000)),
])
tfidf_pipeline.fit(X_train_tf, y_train_tf)
y_pred_tf = tfidf_pipeline.predict(X_val_tf)

# Decision scores via decision_function for ROC/PR
dec_tf = tfidf_pipeline.decision_function(X_val_tf)

results_tf = evaluate_model(
    "TF-IDF + LinearSVC", y_val_tf, y_pred_tf,
    decision_scores=dec_tf, classes=CLASSES, cm_cmap=plt.cm.Blues
)

# Error analysis
error_analysis("TF-IDF + LinearSVC", X_val_tf, y_val_tf, y_pred_tf)

# ── 5-fold cross-validation ───────────────────────────────────────────
print("\n  [5-Fold Cross-Validation — TF-IDF]")
cv_scores = cross_val_score(
    tfidf_pipeline, df_training['cleaned_text_tfidf'], df_training['label'],
    cv=StratifiedKFold(n_splits=5, shuffle=True, random_state=42),
    scoring='f1_weighted'
)
print(f"    CV F1 (weighted) : {cv_scores.mean()*100:.2f}% ± {cv_scores.std()*100:.2f}%")
results_tf['cv_f1_mean'] = cv_scores.mean()
results_tf['cv_f1_std']  = cv_scores.std()

# Retrain on full data for final predictions
tfidf_pipeline.fit(df_training['cleaned_text_tfidf'], df_training['label'])
predictions_tfidf = tfidf_pipeline.predict(df_test['cleaned_text_tfidf'])


# =====================================================================
# 9. MODEL 2: Word2Vec + LinearSVC  (with TF-IDF weighted averaging)
# =====================================================================
print("\n" + "="*60)
print("  MODEL 2: Word2Vec + LinearSVC")
print("="*60)

# Train W2V only on training split (no leakage)
train_w2v_sentences = [s for i in train_idx for s in df_training.loc[i, 'tokens_w2v']]

w2v_model = Word2Vec(
    sentences=train_w2v_sentences,
    vector_size=100, window=5, min_count=2, workers=4, seed=42,
)

# ── TF-IDF weighted document vectors ─────────────────────────────────
# A plain mean treats "the" the same as "nasdaq". Weighting by TF-IDF
# score gives rarer/more informative words more influence in the vector.
tfidf_weights_vectorizer = TfidfVectorizer(max_features=25000)
tfidf_weights_vectorizer.fit([" ".join(t) for t in df_training['tokens_w2v_flat']])

def get_tfidf_weighted_vector(tokens: list, w2v, tfidf_vec) -> np.ndarray:
    """Mean of word vectors weighted by their TF-IDF score in this document."""
    vocab = tfidf_vec.vocabulary_
    idf   = tfidf_vec.idf_
    vec   = np.zeros(w2v.vector_size)
    total_weight = 0.0
    for token in tokens:
        if token in w2v.wv.key_to_index and token in vocab:
            weight       = idf[vocab[token]]
            vec         += weight * w2v.wv[token]
            total_weight += weight
    return vec / total_weight if total_weight > 0 else vec

X_train_w2v = np.array([
    get_tfidf_weighted_vector(df_training.loc[i, 'tokens_w2v_flat'], w2v_model, tfidf_weights_vectorizer)
    for i in train_idx
])
X_val_w2v = np.array([
    get_tfidf_weighted_vector(df_training.loc[i, 'tokens_w2v_flat'], w2v_model, tfidf_weights_vectorizer)
    for i in val_idx
])
y_train_w2v = df_training.loc[train_idx, 'label']
y_val_w2v   = df_training.loc[val_idx,   'label']

w2v_classifier = LinearSVC(C=1.0, class_weight=CLASS_WEIGHT, random_state=42, max_iter=3000)
w2v_classifier.fit(X_train_w2v, y_train_w2v)
y_pred_w2v = w2v_classifier.predict(X_val_w2v)
dec_w2v    = w2v_classifier.decision_function(X_val_w2v)

results_w2v = evaluate_model(
    "Word2Vec + LinearSVC", y_val_w2v, y_pred_w2v,
    decision_scores=dec_w2v, classes=CLASSES, cm_cmap=plt.cm.Greens
)

error_analysis("Word2Vec + LinearSVC", df_training.loc[val_idx, 'cleaned_text_tfidf'], y_val_w2v, y_pred_w2v)

# ── PCA 2-D embedding visualisation ──────────────────────────────────
print("\n  [PCA 2-D — Word2Vec embeddings]")
pca_2d  = PCA(n_components=2, random_state=42)
X_2d    = pca_2d.fit_transform(np.vstack([X_train_w2v, X_val_w2v]))
y_2d    = np.concatenate([y_train_w2v, y_val_w2v])
explained = pca_2d.explained_variance_ratio_

plt.figure(figsize=(8, 6))
scatter = plt.scatter(X_2d[:,0], X_2d[:,1], c=y_2d, cmap='tab10', alpha=0.5, s=12)
plt.colorbar(scatter, label='Class')
plt.xlabel(f'PC1 ({explained[0]*100:.1f}% var)')
plt.ylabel(f'PC2 ({explained[1]*100:.1f}% var)')
plt.title('Word2Vec Embeddings — PCA 2-D', fontweight='bold')
plt.tight_layout()
plt.show()

# Retrain on full data for final predictions
all_w2v_sentences = [s for doc in df_training['tokens_w2v'] for s in doc]
w2v_model_full = Word2Vec(
    sentences=all_w2v_sentences,
    vector_size=100, window=5, min_count=2, workers=4, seed=42,
)
tfidf_weights_vectorizer.fit([" ".join(t) for t in df_training['tokens_w2v_flat']])

X_w2v_all = np.array([
    get_tfidf_weighted_vector(t, w2v_model_full, tfidf_weights_vectorizer)
    for t in df_training['tokens_w2v_flat']
])
w2v_classifier.fit(X_w2v_all, df_training['label'])
X_test_w2v     = np.array([
    get_tfidf_weighted_vector(t, w2v_model_full, tfidf_weights_vectorizer)
    for t in df_test['tokens_w2v_flat']
])
predictions_w2v = w2v_classifier.predict(X_test_w2v)


# =====================================================================
# 10. MODEL 3: Sentence Transformer (all-MiniLM-L6-v2) + LinearSVC
# =====================================================================
print("\n" + "="*60)
print("  MODEL 3: Sentence Transformer (all-MiniLM-L6-v2)")
print("="*60)

st_model = SentenceTransformer('all-MiniLM-L6-v2')
print("Encoding all training sentences...")
X_st_all = st_model.encode(df_training['cleaned_text_trans'].tolist(), show_progress_bar=True)
y_st_all = df_training['label']

# Use same indices as all other models
mask_train = df_training.index.isin(train_idx)
mask_val   = df_training.index.isin(val_idx)
X_train_st, X_val_st = X_st_all[mask_train], X_st_all[mask_val]
y_train_st, y_val_st = y_st_all[mask_train], y_st_all[mask_val]

st_classifier = LinearSVC(C=1.0, class_weight=CLASS_WEIGHT, random_state=42, max_iter=3000)
st_classifier.fit(X_train_st, y_train_st)
y_pred_st = st_classifier.predict(X_val_st)
dec_st    = st_classifier.decision_function(X_val_st)

results_st = evaluate_model(
    "all-MiniLM-L6-v2 + LinearSVC", y_val_st, y_pred_st,
    decision_scores=dec_st, classes=CLASSES, cm_cmap=plt.cm.Purples
)

error_analysis("all-MiniLM-L6-v2", df_training.loc[val_idx, 'cleaned_text_trans'],
               y_val_st, y_pred_st)

# ── PCA 2-D embedding visualisation ──────────────────────────────────
print("\n  [PCA 2-D — Transformer embeddings]")
pca_st  = PCA(n_components=2, random_state=42)
X_st_2d = pca_st.fit_transform(X_st_all)
exp_st  = pca_st.explained_variance_ratio_

plt.figure(figsize=(8, 6))
sc = plt.scatter(X_st_2d[:,0], X_st_2d[:,1], c=y_st_all, cmap='tab10', alpha=0.4, s=10)
plt.colorbar(sc, label='Class')
plt.xlabel(f'PC1 ({exp_st[0]*100:.1f}% var)')
plt.ylabel(f'PC2 ({exp_st[1]*100:.1f}% var)')
plt.title('Transformer Embeddings — PCA 2-D', fontweight='bold')
plt.tight_layout()
plt.show()

# ── 5-fold cross-validation ───────────────────────────────────────────
# Note: encoding is expensive; we reuse already-computed X_st_all.
print("\n  [5-Fold Cross-Validation — Transformer embeddings]")
cv_st = cross_val_score(
    LinearSVC(C=1.0, class_weight=CLASS_WEIGHT, random_state=42, max_iter=3000),
    X_st_all, y_st_all,
    cv=StratifiedKFold(n_splits=5, shuffle=True, random_state=42),
    scoring='f1_weighted'
)
print(f"    CV F1 (weighted) : {cv_st.mean()*100:.2f}% ± {cv_st.std()*100:.2f}%")
results_st['cv_f1_mean'] = cv_st.mean()
results_st['cv_f1_std']  = cv_st.std()

# Retrain on full training data for final predictions
st_classifier.fit(X_st_all, y_st_all)
X_test_st      = st_model.encode(df_test['cleaned_text_trans'].tolist(), show_progress_bar=True)
predictions_st = st_classifier.predict(X_test_st)


# =====================================================================
# 11. FINAL MODEL COMPARISON TABLE
# =====================================================================
print("\n" + "="*60)
print("  FINAL MODEL COMPARISON")
print("="*60)

comparison_df = pd.DataFrame([results_tf, results_w2v, results_st])
comparison_df['accuracy']    = (comparison_df['accuracy']    * 100).round(2)
comparison_df['f1_weighted'] = (comparison_df['f1_weighted'] * 100).round(2)
comparison_df['kappa']       = comparison_df['kappa'].round(4)
comparison_df['roc_auc']     = comparison_df['roc_auc'].round(4)
comparison_df['cv_f1_mean']  = (comparison_df['cv_f1_mean']  * 100).round(2)
comparison_df['cv_f1_std']   = (comparison_df['cv_f1_std']   * 100).round(2)

comparison_df.rename(columns={
    'model':       'Model',
    'accuracy':    'Accuracy (%)',
    'f1_weighted': 'F1 Weighted (%)',
    'kappa':       "Cohen's κ",
    'roc_auc':     'ROC-AUC (macro)',
    'cv_f1_mean':  'CV F1 mean (%)',
    'cv_f1_std':   'CV F1 std (%)',
}, inplace=True)

print(comparison_df.to_string(index=False))

# ── Bar chart comparison ──────────────────────────────────────────────
fig, ax = plt.subplots(figsize=(10, 5))
x     = np.arange(len(comparison_df))
width = 0.22
metrics = ['Accuracy (%)', 'F1 Weighted (%)', 'CV F1 mean (%)']
colors  = ['#4C72B0', '#55A868', '#C44E52']

for j, (metric, color) in enumerate(zip(metrics, colors)):
    vals = comparison_df[metric].values
    bars = ax.bar(x + j*width, vals, width, label=metric, color=color, alpha=0.85)
    for bar in bars:
        ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.3,
                f'{bar.get_height():.1f}', ha='center', va='bottom', fontsize=8)

ax.set_xticks(x + width)
ax.set_xticklabels(comparison_df['Model'], rotation=10, ha='right')
ax.set_ylim(0, 115)
ax.set_ylabel('Score (%)')
ax.set_title('Model Comparison — Accuracy / F1 / Cross-Validated F1', fontweight='bold')
ax.legend()
plt.tight_layout()
plt.show()


# =====================================================================
# 12. SAVING FINAL RESULTS
# =====================================================================
print("\nSaving final test data results...")

# str[1:] strips the original leading label char so we can prepend the predicted label.
# Assumes news_column format: '<label><space><text>' (same format as training data).
df_test['final_tfidf']       = predictions_tfidf.astype(str) + df_test['news_column'].str[1:]
df_test['final_w2v']         = predictions_w2v.astype(str)   + df_test['news_column'].str[1:]
df_test['final_transformer'] = predictions_st.astype(str)    + df_test['news_column'].str[1:]

df_test['final_tfidf'].to_csv(
    "dataset/testing_data_predicted_tfidf.csv",
    index=False, header=False, quoting=csv.QUOTE_NONE, escapechar='\\'
)
df_test['final_w2v'].to_csv(
    "dataset/testing_data_predicted_w2v.csv",
    index=False, header=False, quoting=csv.QUOTE_NONE, escapechar='\\'
)
df_test['final_transformer'].to_csv(
    "dataset/testing_data_predicted_transformer.csv",
    index=False, header=False, quoting=csv.QUOTE_NONE, escapechar='\\'
)

# Save comparison table to CSV for easy reference
comparison_df.to_csv("dataset/model_comparison.csv", index=False)

print("\n✅ ALL DONE! Files saved:")
print("   • dataset/testing_data_predicted_tfidf.csv")
print("   • dataset/testing_data_predicted_w2v.csv")
print("   • dataset/testing_data_predicted_transformer.csv")
print("   • dataset/model_comparison.csv")
```
